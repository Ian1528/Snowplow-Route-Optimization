```python
import networkx as nx
import random
import matplotlib.pyplot as plt



# Create a multigraph
multigraph = nx.MultiGraph()

# Add three nodes
multigraph.add_nodes_from([1, 2, 3])

# Add two edges between each pair of nodes
multigraph.add_edges_from([(1, 2), (1, 2), (2, 3), (2, 3), (3, 1), (3, 1)])
# Assign random weights to each edge
random.seed(0)

pos = nx.spring_layout(multigraph) 
nx.draw(multigraph, pos, with_labels=True) 
plt.show()

# Print the multigraph
for u, v, key, data in multigraph.edges(keys=True, data=True):
    data['weight'] = random.choice([1,2,3])
```

```python
for edge in multigraph.edges(data=True):
    print(edge)

```

```python
predecessors, distances = nx.floyd_warshall_predecessor_and_distance(multigraph,weight="distance")
print(distances)
path = nx.reconstruct_path(1, 3, predecessors)
print(path)
```

To reconstruct a path in a multigraph, you have to manually find the shortest edge between nodes in the path
