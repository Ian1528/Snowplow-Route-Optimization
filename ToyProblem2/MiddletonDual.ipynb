# A small chunk of Middleton's Streets

```python
import geopandas
import matplotlib.pyplot as plt
import momepy
import osmnx as ox
import networkx as nx
from contextily import add_basemap
from libpysal import weights
```

```python
x, y = 43.095605631640105, -89.50925942968578

G = ox.graph_from_point((x, y), dist=300, network_type="drive")
```

```python
ox.plot_graph(G)
```

```python
gdf = momepy.nx_to_gdf(G)
gdf
graph = momepy.gdf_to_nx()
```

```python
# Read in example street network from GeoPackage
streets = gdf

# Construct the primal graph
G_primal = momepy.gdf_to_nx(gdf)

# Plot
f, ax = plt.subplots(1, 2, figsize=(12, 6), sharex=True, sharey=True)
streets.plot(color="k", ax=ax[0])
for i, facet in enumerate(ax):
    facet.set_title(("Streets", "Graph")[i])
    facet.axis("off")
    add_basemap(facet)
nx.draw(
    G_primal, {n: [n[0], n[1]] for n in list(G_primal.nodes)}, ax=ax[1], node_size=50
)
```
